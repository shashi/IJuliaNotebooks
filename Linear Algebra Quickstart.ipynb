# Add interaction to your plots

You can use the `Interact` package to create interactive plots with the `@manipulate` macro. Here, we are plotting the Beta distribution from the `Distributions` package, where $\alpha$ and $\beta$ are parameters that can be varied between 1 to 100 using sliders!

There are a few features of Julia that have gone in to this little snippet of code, namely macros, anonymous functions, and unicode support. We do not explain all these ideas here, but only try to leave you with a taste of what more can be done with Julia.

```julia
using Interact, Gadfly, Distributions

@manipulate for α in 1:100, β = 1:100
    plot(x -> pdf(Beta(α, β), x), 0, 1)
end
```

# Three ways to do a three-point stencil

In this example we show that Julia allows the same code to be written with trade-offs between brevity, speed, and memory consumption.

1. Vectorized implementation using array indexing that shifts the array left and right. This tends to be clear but allocates a lot more memory.
2. De-vectorized implementation using a `for` loop. This is more verbose, but the fastest.
3. The same operation using a sparse matrix-vector multiplication. This implementation uses a sparse-matrix formulation. While construction of the sparse matrix takes time and memory, the multiplication is fast. This formulation is sometimes more convenient for irregular problems arising in applications.

```julia
# Method 1

# Here we are defining a single-line function stencilvec which takes A as input
stencilvec(A) = -A[3:end]./4 + 2A[2:end-1] - A[1:end-2]./4
```

```julia
# Method 2

# This is a multi-line function
# Note that blocks in Julia are delimited by end
function stencilloop(A)
    n = length(A)
    B = zeros(eltype(A), n) # an array of n zeros of the same type as elements of A
    for i=2:n-1
        B[i] = -A[i+1]/4 + 2A[i] - A[i-1]/4
    end
    B # return B
end
```

```julia
# Method 3

function stencilmatvec(A)
    n = length(A)
    S = spdiagm((-ones(n-1)./4, 2*ones(n), -ones(n-1)./4), (-1, 0 , 1)) # Construct a sparse matrix from diagonals
    S*A
end
```

```julia
A = ones(1000000)
@time stencilvec(A)
@time stencilloop(A)
@time stencilmatvec(A);
```

# Eigshow

```julia

Gadfly.set_default_plot_size(5inch, 5inch)
A = [1 3; 4 2] / 4

@manipulate for x=slider(0:.01:2π, value=0., label="x")
    xs = [cos(0:0.05:x) sin(0:0.05:x)]'
    Axs = A*xs
    plot(layer(x=xs[1, :], y=xs[2, :],   Geom.path),
         layer(x=Axs[1, :], y=Axs[2, :], Geom.path),
         Scale.x_continuous(minvalue=-2, maxvalue=2),
         Scale.y_continuous(minvalue=-2, maxvalue=2))
end
```
