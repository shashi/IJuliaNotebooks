[Gadfly](http://gadflyjl.org/) is a statical gaphics package written entirely in Julia. It is much-loved for its stunning plots. Gadfly is based on Leland Wilkinson's book [The Grammar of Graphics](http://www.cs.uic.edu/~wilkinson/TheGrammarOfGraphics/GOG.html).

```julia
using Gadfly
```

Anatomy of a plot. Every Gadfly plot can be configured using the following components, passed to the `plot` function.

* **Aesthetics**: parts of a plot that can be bound to data. `x` and `y` are aesthetics in the usual scenarios. They define where x and y values for the plot come from.
* **Geometry**: The geometry of the plot. Usually defines the kind of plot to use, some geoms take arguments for further configuration. Geometry can contain in itself aesthetics (data bindings).
* **Scale**: Configuration for scales on either axis. e.g. `Scale.y_log` makes the scale on the y axis logarithmic.
* **Guides**: Configuration for labels, ticks, title, and colors

```julia
# plot 10 random points:

x_values = [1:100]
y_values = cumsum(randn(100)) # cumulative sum of n normal random numbers, aka a random walk

plot(x=x_values, y=y_values) # x and y are aesthetics
```

```julia
# plot a line-plot of the same values
plot(x=x_values, y=y_values, Geom.line, Guide.title("random line plot")) # Geom.line selects the line geometry
# Guide.title sets the title
```

```julia
# show both points and lines!
plot(x=x_values,
     y=y_values,
     Geom.line, 
     Geom.point,  # Geom.point and Geom.line turn on both point and line geometry
     Guide.title("A random walk"),
     Guide.xlabel("Iterations"),
     Guide.ylabel("Position")
)
```

## Plotting functions

Continuous functions can be plotted by passing the fuction as the first argument, and the x-axis range as the second argument.

```julia
f(x) = e^x


plot(f, 0, 12)
```

```julia
p1 = plot(f, 0, 12)
p2 = plot(f, 0, 12, Scale.y_log) # Other components can be added to this plot. eg y axis on a log-scale.

hstack(p1, p2) # Stack the plots horizontally
```

You can also pass a vector of functions to plot all of them simultaneously

```julia
plot([sin, cos, floor], 0, 12)
```

```julia
wave(x, y) = x*exp(-(x-int(x))^2-y^2)
plot(z=wave, x=linspace(-8,8,150), y=linspace(-2,2,150), Geom.contour)
```
